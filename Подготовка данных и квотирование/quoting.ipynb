```python
# === CELL 1: helpers (просто вставь в ноутбук) ===

def _z_test_two_proportions(x1: int, n1: int, x2: int, n2: int, alternative: str = "two-sided"):
    """
    Z-тест разницы долей (pooled).
    alternative: "two-sided" | "larger" | "smaller"
    Возвращает (z, p).
    """
    if n1 == 0 or n2 == 0:
        return np.nan, np.nan

    p_pool = (x1 + x2) / (n1 + n2)
    denom = np.sqrt(p_pool * (1 - p_pool) * (1 / n1 + 1 / n2))
    if denom == 0:
        return np.nan, np.nan

    z = (x1 / n1 - x2 / n2) / denom

    if alternative == "two-sided":
        p = 2 * scipy.stats.norm.sf(np.abs(z))
    elif alternative == "larger":     # H1: p1 > p2
        p = scipy.stats.norm.sf(z)
    elif alternative == "smaller":    # H1: p1 < p2
        p = scipy.stats.norm.cdf(z)
    else:
        raise ValueError("alternative must be: two-sided|larger|smaller")

    return z, p


def _gender_z_test_drop_minus1(df_a: pd.DataFrame, df_b: pd.DataFrame, col: str = "socdec_gender"):
    """
    Z-тест разницы долей по полу.
    ВАЖНО: значения -1 удаляем ТОЛЬКО в рамках этого теста (локально).
    Сравниваем долю "Мужской" (=1) между A и B.
    """
    a = pd.to_numeric(df_a[col], errors="coerce")
    b = pd.to_numeric(df_b[col], errors="coerce")

    a = a.dropna()
    b = b.dropna()

    a = a[a != -1]
    b = b[b != -1]

    n1 = int(a.shape[0])
    n2 = int(b.shape[0])

    x1 = int((a == 1).sum())
    x2 = int((b == 1).sum())

    z, p = _z_test_two_proportions(x1, n1, x2, n2, alternative="two-sided")

    p1 = (x1 / n1) if n1 else np.nan
    p2 = (x2 / n2) if n2 else np.nan

    return z, p, n1, n2, p1, p2


def _mann_whitney_test(df_a: pd.DataFrame, df_b: pd.DataFrame, col: str, alternative: str = "two-sided"):
    """
    Mann–Whitney U (две независимые выборки).
    """
    x = pd.to_numeric(df_a[col], errors="coerce").dropna()
    y = pd.to_numeric(df_b[col], errors="coerce").dropna()

    n1 = int(x.shape[0])
    n2 = int(y.shape[0])

    if n1 == 0 or n2 == 0:
        return np.nan, np.nan, n1, n2

    u, p = scipy.stats.mannwhitneyu(x, y, alternative=alternative)
    return u, p, n1, n2


def _chi2_test(df_a: pd.DataFrame, df_b: pd.DataFrame, col: str):
    """
    Chi-square test of independence:
    (dataset A vs dataset B) x (категории col)
    Категорий может быть сколько угодно — не надо заранее знать число.
    Возвращает chi2, p, dof, min_expected, contingency_table
    """
    a = df_a[[col]].copy()
    b = df_b[[col]].copy()

    a["_dataset_"] = "A"
    b["_dataset_"] = "B"
    both = pd.concat([a, b], ignore_index=True)

    both = both.dropna(subset=[col])

    # нормализуем типы/пробелы, иначе категории могут раздваиваться
    s = both[col].astype("object")
    s = s.apply(lambda v: v.strip() if isinstance(v, str) else v)
    both[col] = s.astype("object").astype(str)

    ct = pd.crosstab(both["_dataset_"], both[col])

    if ct.shape[0] < 2 or ct.shape[1] < 2:
        return np.nan, np.nan, np.nan, np.nan, ct

    chi2, p, dof, expected = scipy.stats.chi2_contingency(ct)
    min_expected = float(np.min(expected)) if expected is not None else np.nan
    return chi2, p, dof, min_expected, ct


def run_pairwise_tests(pyaterka: pd.DataFrame, magnit: pd.DataFrame, lenta: pd.DataFrame):
    """
    Тесты по правилам, которые ты задал:
    - socdec_gender: z для доли (локально выкидываем -1)
    - socdec_age_years: Mann–Whitney U
    - socdec_primary_shopper: chi-square
    - socdec_household_size: chi-square
    - socdec_material_status: chi-square
    - visit_freq: chi-square
    """
    pairs = [
        ("pyaterka", pyaterka, "magnit", magnit),
        ("pyaterka", pyaterka, "lenta",  lenta),
        ("magnit",   magnit,   "lenta",  lenta),
    ]

    plan = [
        ("socdec_gender", "z_prop"),
        ("socdec_age_years", "mannwhitney"),
        ("socdec_primary_shopper", "chi2"),
        ("socdec_household_size", "chi2"),
        ("socdec_material_status", "chi2"),
        ("visit_freq", "chi2"),
    ]

    rows = []

    for a_name, a_df, b_name, b_df in pairs:
        for col, test in plan:
            if col not in a_df.columns or col not in b_df.columns:
                rows.append({
                    "pair": f"{a_name} vs {b_name}",
                    "variable": col,
                    "test": test,
                    "stat": np.nan,
                    "p_value": np.nan,
                    "n_a": np.nan,
                    "n_b": np.nan,
                    "extra": "column_missing"
                })
                continue

            if test == "z_prop":
                z, p, n1, n2, p1, p2 = _gender_z_test_drop_minus1(a_df, b_df, col=col)
                rows.append({
                    "pair": f"{a_name} vs {b_name}",
                    "variable": col,
                    "test": "z_test_proportion(male=1)",
                    "stat": z,
                    "p_value": p,
                    "n_a": n1,
                    "n_b": n2,
                    "extra": f"p_male_a={p1:.4f} | p_male_b={p2:.4f} | dropped(-1) локально"
                })

            elif test == "mannwhitney":
                u, p, n1, n2 = _mann_whitney_test(a_df, b_df, col=col)
                rows.append({
                    "pair": f"{a_name} vs {b_name}",
                    "variable": col,
                    "test": "Mann–Whitney U",
                    "stat": u,
                    "p_value": p,
                    "n_a": n1,
                    "n_b": n2,
                    "extra": ""
                })

            elif test == "chi2":
                chi2, p, dof, min_exp, ct = _chi2_test(a_df, b_df, col=col)
                rows.append({
                    "pair": f"{a_name} vs {b_name}",
                    "variable": col,
                    "test": "Chi-square",
                    "stat": chi2,
                    "p_value": p,
                    "n_a": int(a_df[col].notna().sum()),
                    "n_b": int(b_df[col].notna().sum()),
                    "extra": f"dof={dof} | min_expected={min_exp:.3f} | k={ct.shape[1]}"
                })
            else:
                raise ValueError(f"Unknown test type: {test}")

    res = pd.DataFrame(rows)

    # удобная сортировка: сначала пара, потом p-value
    res = res.sort_values(["pair", "p_value"], ascending=[True, True], na_position="last").reset_index(drop=True)
    return res
```

```python
def categorical_frequencies(df: pd.DataFrame, df_name: str, cols: list[str], normalize: bool = False):
    """
    Печатает распределения частот по категориальным колонкам.
    normalize=False -> абсолютные частоты
    normalize=True  -> доли
    """
    print(f"\n==================== {df_name.upper()} ====================")

    for col in cols:
        if col not in df.columns:
            print(f"\n[{col}] — НЕТ В ДАТАСЕТЕ")
            continue

        print(f"\n[{col}]")
        vc = (
            df[col]
            .astype("object")
            .apply(lambda v: v.strip() if isinstance(v, str) else v)
            .value_counts(dropna=False, normalize=normalize)
        )

        display(vc.to_frame("freq" if not normalize else "share"))
```

```python
def relative_freq_table_3dfs(
    pyaterka: pd.DataFrame,
    magnit: pd.DataFrame,
    lenta: pd.DataFrame,
    col: str
):
    """
    Возвращает DataFrame:
    index   -> категории переменной col
    columns -> pyaterka | magnit | lenta
    values  -> относительные частоты (доли)
    
    Если категории нет в датасете — ставится 0.
    """
    def _rel_freq(df: pd.DataFrame):
        if col not in df.columns:
            return pd.Series(dtype=float)

        s = (
            df[col]
            .astype("object")
            .apply(lambda v: v.strip() if isinstance(v, str) else v)
        )

        return s.value_counts(normalize=True, dropna=False)

    p = _rel_freq(pyaterka)
    m = _rel_freq(magnit)
    l = _rel_freq(lenta)

    table = pd.concat(
        [p, m, l],
        axis=1,
        keys=["pyaterka", "magnit", "lenta"]
    ).fillna(0)

    # для красоты: сортировка категорий по суммарной доле
    table = table.loc[
        table.sum(axis=1).sort_values(ascending=False).index
    ]

    return table
```

```python
from itertools import combinations

datasets = {
    "pyaterka": pyaterka,
    "magnit": magnit,
    "lenta": lenta,
}

for (name_a, df_a), (name_b, df_b) in combinations(datasets.items(), 2):
    print("\n" + "=" * 80)
    print(f"CHI-SQUARE (без <18): socdec_age_cat — {name_a} vs {name_b}")
    print("=" * 80)

    # берем только возрастную категорию
    a = df_a[["socdec_age_cat"]].copy()
    b = df_b[["socdec_age_cat"]].copy()

    a["dataset"] = name_a
    b["dataset"] = name_b

    both = pd.concat([a, b], ignore_index=True)

    # локально убираем <18 и NaN
    both = both[
        both["socdec_age_cat"].notna() &
        (both["socdec_age_cat"] != "<18")
    ]

    # таблица сопряженности
    ct = pd.crosstab(both["dataset"], both["socdec_age_cat"])

    display(ct)

    # защита от вырожденных случаев
    if ct.shape[0] < 2 or ct.shape[1] < 2:
        print("χ² не проводится: недостаточно категорий")
        continue

    chi2, p, dof, expected = scipy.stats.chi2_contingency(ct)

    print(f"chi2 = {chi2:.3f}")
    print(f"dof  = {dof}")
    print(f"p-value = {p:.5f}")
    print(f"min expected = {expected.min():.3f}")
```

```python
# -----------------------------
# настройки квотирования
# -----------------------------
quota_cols = [
    "socdec_material_status",
    "socdec_gender",
    "socdec_primary_shopper",
    "socdec_age_cat",
]

datasets = {
    "pyaterka": pyaterka,
    "magnit": magnit,
    "lenta": lenta,
}

rng = np.random.default_rng(42)  # фиксируем сид для воспроизводимости


# -----------------------------
# 1. оставляем только строки, где ВСЕ квотные признаки заполнены
# -----------------------------
cleaned = {}
for name, df in datasets.items():
    cleaned[name] = df.dropna(subset=quota_cols).copy()


# -----------------------------
# 2. считаем размер каждой страты в каждом датасете
# -----------------------------
group_sizes = {}

for name, df in cleaned.items():
    group_sizes[name] = (
        df
        .groupby(quota_cols)
        .size()
        .rename(name)
    )

# объединяем в одну таблицу
sizes_df = pd.concat(group_sizes.values(), axis=1).fillna(0).astype(int)

# -----------------------------
# 3. целевая квота = минимум по трем сетям в каждой страте
# -----------------------------
sizes_df["target_n"] = sizes_df.min(axis=1)

# убираем страты, где хотя бы в одной сети 0 наблюдений
sizes_df = sizes_df[sizes_df["target_n"] > 0]

print("Количество страт, попавших в квотирование:", sizes_df.shape[0])
display(sizes_df.head())
```

```python
# -----------------------------
# 4. случайно отбираем target_n наблюдений в каждой страте
# -----------------------------
balanced = {}

for name, df in cleaned.items():
    parts = []

    for strata_values, row in sizes_df.iterrows():
        target_n = row["target_n"]

        # фильтр по конкретной страте
        mask = np.ones(len(df), dtype=bool)
        for col, val in zip(quota_cols, strata_values):
            mask &= (df[col] == val)

        sub = df.loc[mask]

        # случайная подвыборка
        sampled_idx = rng.choice(sub.index, size=target_n, replace=False)
        parts.append(df.loc[sampled_idx])

    balanced[name] = pd.concat(parts, ignore_index=True)

    print(f"{name}: было {len(df)}, стало {len(balanced[name])}")
```
